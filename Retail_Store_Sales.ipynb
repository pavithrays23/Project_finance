```python
import pandas as pd
import numpy as np

# Load the CSV file into Python
df = pd.read_csv('Sample - Superstore (1).csv', encoding='latin1')
```

```python
# Show the first 5 rows of data
df.head()
```

```python
# Show the first 5 rows of data
display(df.head())
```

```python
# Information about data (columns, datatypes)
df.info()

# Summary statistics (like average, min, max)
df.describe()
```

```python
# Check for missing values
df.isnull().sum()

# Check for duplicate rows
df.duplicated().sum()
```

```python
# Remove duplicate rows
df = df.drop_duplicates()

# Remove rows with missing values (if any)
df = df.dropna()

# Convert 'Order Date' to proper date format
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
# Save cleaned data to a new CSV file
df.to_csv('Cleaned_Superstore.csv', index=False)

# Download the cleaned file
files.download('Cleaned_Superstore.csv')
```

```python
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
region_sales.plot(kind='bar')
```

```python
category_sales = df.groupby(['Category', 'Sub-Category'])[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False)
category_sales.plot(kind='barh', figsize=(10,8))
```

```python
import seaborn as sns
sns.scatterplot(x='Discount', y='Profit', data=df)
```

```python
segment_profit = df.groupby('Segment')['Profit'].sum()
segment_profit.plot(kind='pie', autopct='%1.1f%%')
```

```python
df['Profit_Margin_%'] = (df['Profit'] / df['Sales']) * 100
```

```python
top_products = df.groupby('Product Name')['Profit'].sum().sort_values(ascending=False).head(10)
```
